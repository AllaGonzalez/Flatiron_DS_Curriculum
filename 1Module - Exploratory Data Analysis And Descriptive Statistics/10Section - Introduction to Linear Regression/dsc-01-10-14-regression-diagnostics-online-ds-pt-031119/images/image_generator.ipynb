```python
import numpy as np
import scipy.stats as stats


import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 2))

x3 = np.random.gamma(1, scale=1, size=10**5)

ax1 = plt.subplot(121)
stats.probplot(x3, dist="norm", plot=ax1);

ax2 = plt.subplot(122)
sns.kdeplot(x3, ax=ax2)


```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 2))

x3 = np.random.gamma(0.0001, scale=2, size=10**5)

ax1 = plt.subplot(121)
stats.probplot(x3, dist="norm", plot=ax1);

ax2 = plt.subplot(122)
sns.kdeplot(x3, ax=ax2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 2))

x2 = list(np.random.normal(loc=0, scale=3, size=10**4)) + list(np.random.rand(3*10**3)) + list(-1*np.random.rand(3*10**3))
ax1 = plt.subplot(121)
stats.probplot(x2, dist="norm", plot=ax1);

ax2 = plt.subplot(122)
sns.kdeplot(x2, ax=ax2)
```

## All Together

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(18, 8))

#Skewed Right
x1 = np.random.gamma(1, scale=1, size=10**5)

ax1 = plt.subplot(421)
stats.probplot(x1, dist="norm", plot=ax1);
ax1.set_title('QQ Plot of X1')

ax2 = plt.subplot(422)
sns.kdeplot(x1, ax=ax2)
ax2.set_title('Density Plot of X1')

#Skewed Left (Weibull)
x2 = np.random.weibull(18, size=10**5)

ax3 = plt.subplot(423)
stats.probplot(x2, dist="norm", plot=ax3);
ax3.set_title('QQ Plot of X2')

ax4 = plt.subplot(424)
sns.kdeplot(x2, ax=ax4)
ax4.set_title('Density Plot of X2')

#Peaked
x3 = list(np.random.normal(loc=0, scale=3, size=10**4)) + list(np.random.rand(3*10**3)) + list(-1*np.random.rand(3*10**3))
ax5 = plt.subplot(425)
stats.probplot(x3, dist="norm", plot=ax5);
ax5.set_title('QQ Plot of X3')

ax6 = plt.subplot(426)
sns.kdeplot(x3, ax=ax6)
ax6.set_title('Density Plot of X3')

#Triangular
x4 = np.random.triangular(-2,0,2, 10**5)
ax7 = plt.subplot(427)
stats.probplot(x4, dist="norm", plot=ax7);
ax7.set_title('QQ Plot of X4')

ax8 = plt.subplot(428)
sns.kdeplot(x4, ax=ax8)
ax8.set_title('Density Plot of X4')

plt.tight_layout();
```
