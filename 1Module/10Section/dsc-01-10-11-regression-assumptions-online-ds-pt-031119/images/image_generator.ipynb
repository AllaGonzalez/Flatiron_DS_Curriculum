```python
import numpy as np
import scipy.stats as stats


import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

## Histograms

```python
x1 = np.random.normal(size=200)
x2 = np.array(list(np.random.uniform(low=-1, high=0, size=180)) + list(np.random.uniform(low=0, high=4, size=20)))
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(12,6))

ax1 = plt.subplot(121)
ax1_label = "Std= {}\nMean= {}\nN= {}".format(round(x1.std(), 2), round(x1.mean(), 2), len(x1))
sns.distplot(x1, ax=ax1, label=ax1_label)
ax1.set_title('Normally Distributed Residuals')
ax1.set_xlabel('Normalized Residuals')
ax1.set_ylabel('Frequency')
ax1.legend(loc=(1,0))

ax2 = plt.subplot(122)
ax2_label = "Std= {}\nMean= {}\nN= {}".format(round(x2.std(), 2), round(x2.mean(), 2), len(x2))
sns.distplot(x2, ax=ax2, label=ax2_label)
ax2.set_title('Problematic Residuals: Model Assumptions Violated')
ax2.set_xlabel('Normalized Residuals')
ax2.set_ylabel('Frequency')
ax2.legend(loc=(1,0))

plt.tight_layout()
```

## QQ Plots

```python
fig, axes = plt.subplots(ncols=2, nrows=3, figsize=(12, 18))

#Normal
x1 = np.random.normal(size=10**4)
ax1 = plt.subplot(321)
sns.distplot(x1, ax=ax1)
ax1.set_title('The Standard Normal Distribution');

ax2 = plt.subplot(322)
stats.probplot(x1, dist="norm", plot=ax2);

#Peaked
x2 = list(np.random.normal(loc=0, scale=3, size=10**4)) + list(np.random.rand(2*10**4)) + list(-1*np.random.rand(2*10**4))
ax3 = plt.subplot(323)
sns.distplot(x2, ax=ax3)
ax3.set_title('Nonormal; Data too Peaked in Middle');

ax4 = plt.subplot(324)
stats.probplot(x2, dist="norm", plot=ax4);

#Skewed
x3 = np.random.gamma(1, scale=1, size=10**5)
ax5 = plt.subplot(325)
sns.distplot(x3, ax=ax5)
ax5.set_title('A Skewed Right Distribution');

ax6 = plt.subplot(326)
stats.probplot(x3, dist="norm", plot=ax6);
axes[2][1].set_title('Normal QQ Plot');
```
